3. Preparación de los datos:

- Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
# la funcion la creamos incorporando isinstances para verificar que el value es de tipo string y añadimos un for loop para iterar en las columnas
# y me aplique la función en todas
def remplazar(value):
        if isinstance(value, str):
            try:
                return float(value.replace(',', '.'))
            except ValueError:
                return value
        return value

for column in df_spotify:
    df_spotify[column] = df_spotify[column].apply(remplazar)
```

```python
df_spotify.head(1)
```

```python
# hacemos la misma operación que en el ejercicio anterior pero haciendo la media para las columnas visualizaciones y likes.
df_visitas_likes = df_spotify.groupby("artist")['views', 'likes'].mean()
df_visitas_likes.reset_index(inplace = True)
df_visitas_likes
```

```python
df_mas_visitas = df_visitas_likes.sort_values(by="views", ascending=False)
df_visitas = df_mas_visitas.drop(['likes'], axis=1)
df_visitas.head(10)
```

```python
df_visitas.dtypes
```

- Quedaros con los 10 artistas que más likes han tenido.

```python
df_mas_likes = df_visitas_likes.sort_values(by="likes", ascending=False)
df_likes = df_mas_likes.drop(['views'], axis=1)
df_likes.head(10)
```

- Haced un subplot, donde en uno tengamos los 10 artistas con más likes y en el otro los 10 artistas con mas views

```python
# importamos la librería Matplotlib antes de nada, al principio del notebook
```

```python
fig,ax = plt.subplots(2,1, figsize=(10,8)) #esta será la gráfica de empleados por departamento y puestos, con las instrucciones del enunciado

ax[0].bar(df_visitas["views"], df_visitas["artist"], edgecolor='black', color='grey')
ax[0].set_xlabel('Views')
ax[0].set_ylabel('Artistas')

ax[1].bar(df_likes['likes'], df_likes['artist'], edgecolor='black', color='pink')
ax[1].set_xlabel('Views')
ax[1].set_ylabel('Artistas')

ax[0].bar(df_visitas['views'], df_visitas['artist'].apply(str))
ax[1].bar(df_likes['likes'], df_likes['artist'] )


ax[0].set_title('Visualizaciones')
ax[1].set_title('Likes')

plt.show()
```

Para esto tendréis que:
    - Filtrad los datos para quedaros solo con aquellos datos donde en la columna "track" tengamos "feat".

```python
df_spotify.head(2)
```

```python
df_spotify['track'].unique()
```

```python
df_spotify['track'].isna().sum()
```

```python
patron = 'feat.'
```

```python
# he tenido que buscar elmétodo máscara booleana, porque me estaba dando error como si hubieran valores faltantes, aunque previamente he verificado con isna y no tiene valores faltantes.
# No obstante, hemos aplicado la máscara al str.contains y nos saca las track que contienen feat.

mask = df_spotify['track'].isna()

df_track = df_spotify.loc[~mask & df_spotify['track'].str.contains('feat'), :]
df_track.head(2)
```

Contad cuantas canciones tiene cada artista utilizando el dataframe generado en el paso anterior.

```python
# contamos los valores por cada artista

canciones = df_track.groupby('artist')['track'].value_counts()
canciones
```

```python
# y sumamos los registros por cada artista

canciones_cantante = canciones.groupby('artist').sum()
canciones_cantante
```

- Mostrad los resultados en una gráfica, seleccionado los 10 artistas con más colaboraciones.

```python
# primero ordenamos los valores del df canciones_cantante
canciones_cantante_ordenado = canciones_cantante.sort_values(ascending=False)
canciones_cantante_ordenado
```

```python
# y lo guardamos en un df para poder sacar la grafica

df_canciones_cantante_ordenado = pd.DataFrame(canciones_cantante_ordenado).reset_index()
df_mas_canciones = df_canciones_cantante_ordenado.head(10)
df_mas_canciones
```

```python
plt.figure(figsize=(8,4))
fig = sns.barplot(x='track', y='artist', data=df_mas_canciones)
fig.set(xlabel='Canciones', ylabel='Artistas')
fig.set_title('Colaboraciones artistas')

plt.show()
```
